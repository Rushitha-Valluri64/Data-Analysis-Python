```python
#Sclar constants
a=tf.constant(42)
b=tf.constant(3.14)
c=tf.constant("Hello Tensorflow")
print(a)
print(a.numpy())
print(b)
print(b.numpy())
print(c)
print(c.numpy())
```

```python
#Vector constants
a=tf.constant([1,2,3])
b=tf.constant([1.0,2.0,3.0])
print(a)
print(a.numpy())
print(b)
print(b.numpy())
```

```python
#Matrix constants
a=tf.constant([[1,2],[3,4]])
b=tf.constant([[1.0,2.0],[3.0,4.0]])
print(a)
print(a.numpy())
print(b)
print(b.numpy())
```

```python
#Higher dimension constants
a=tf.constant([[[1,2,0],[3,4,0]],[[5,6,0],[7,8,0]]])
print(a)
print(a.numpy())
```

```python

#Boolean constants
a=tf.constant(True)
print(a)
```

```python
#Special constants
a=tf.zeros(shape=(2,2))
b=tf.ones(shape=(3,3))
print(a)
print(a.numpy())
print(b)
print(b.numpy())
```

```python
import tensorflow as tf
tf.config.run_functions_eagerly(True)
a=tf.constant(5.0)
b=tf.constant(3.0)
c=tf.multiply(a,b)
result=a*b
d=a@b
print(result)
print(result.numpy())
print(d)
print(d.numpy())
```

```python
import tensorflow as tf
a=tf.constant(1.25)
print(a)
print(a.numpy())
b=tf.cast(a,dtype=tf.int32)
print(b)
print(b.numpy())
```

```python
import tensorflow as tf
x=tf.constant([[1.,2.,3.],[4.,5.,6.]])
print(x)
print(x.numpy())

e=tf.transpose(x)
print(e.numpy())

r=tf.reduce_sum(x)
print(r.numpy())
```

```python
x=tf.constant([[1.,2.,3.],[4.,5.,6.]])
j=x @ tf.transpose(x)
print(j.numpy())
f=tf.concat([x,x,x],axis=0)
print(f.numpy())
```
