```python
import pandas as pd
from fpdf import FPDF
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv('/content/drive/MyDrive/wtsup/Akka.csv')
print(df)
```

```python
# Convert DataFrame to a string
df_str = df.to_string(index=False)
```

```python
# Initialize the PDF
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)
```

```python
# Add the DataFrame content to the PDF
# Encode the string using UTF-8 to avoid UnicodeEncodeError
pdf.multi_cell(0, 10, df_str.encode('utf-8').decode('latin-1'))
```

```python
# Save the PDF file
pdf.output("/content/WhatsApp_Chat.pdf")

print("PDF successfully created!")
```
