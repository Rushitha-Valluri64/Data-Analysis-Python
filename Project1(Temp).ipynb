```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_excel("/content/Expected Temparatures.xlsx")
print(df)
average_temperature =df['Temparature'].mean()
highest_temperature = df['Temparature'].max()
lowest_temperature = df['Temparature'].min()
threshold=19
days_exceeding_threshold = df[df['Temparature'] > threshold].shape[0]
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Temparature'], color='blue', marker='o')
plt.title('Daily Temperatures')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.axhline(y=threshold, color='red', linestyle='--', label=f'Threshold ({threshold}C)')
plt.legend()
plt.grid(True)
plt.show()

# Print insights
print("Average temperature:", average_temperature, "C")
print("Highest temperature:", highest_temperature, "C")
print("Lowest temperature:", lowest_temperature, "C")
print("Number of days where temperature exceeded", threshold, "C:", days_exceeding_threshold)
```
