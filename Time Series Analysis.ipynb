```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate sample time series data
np.random.seed(0)
time = np.arange(0, 10, 0.1)
data = np.sin(time) + np.random.normal(scale=0.1, size=len(time))
```

```python
# Plot the time series data
plt.plot(time, data, label='Time Series Data',color='hotpink')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series Data Analysis')
plt.legend(loc="best")
plt.show()
```

```python
# Compute statistical metrics
mean_value = np.mean(data)
std_dev = np.std(data)
autocorr = np.correlate(data, data, mode='full') / len(data)

print("Mean:", mean_value)
print("Standard Deviation:", std_dev)
print("Auto correlation:", autocorr)
```
