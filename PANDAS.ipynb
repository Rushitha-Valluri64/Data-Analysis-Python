```python
import matplotlib.pyplot as plt
a=[230,560,780,127,128]
b=[200,160,270,127,400]
years=[1,2,3,4]
profit_a=[(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b=[(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color="hotpink",linewidth= '2',label="Company A",marker='.',ms='10',mec='m')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='blue',linestyle='dotted',label='Company B',marker='.')
```
